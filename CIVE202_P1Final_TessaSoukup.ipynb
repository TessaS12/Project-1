```python
#import and frame data
import pandas as pd
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
air_data = pd.DataFrame(air_data)
```

#Task 1

```python
#grouping measurements by location, and finding mean and median
concentration_summary = air_data.groupby(['sensor.name']).agg({
                    'voc': ['mean','median'],
                    'pm2.5_atm': ['mean','median'],
                    'pm10.0_atm': ['mean','median']})
#isolating VOC and printing
highest_voc_concentration = concentration_summary.sort_values(by = ('voc', 'mean'), ascending = False)
highest_voc_concentration = highest_voc_concentration[['voc']]
highest_voc_concentration.head(5)
```

```python
#isolating and printing top 5 pm2.5 locations
highest_pm25_concentration = concentration_summary.sort_values(by = ('pm2.5_atm', 'mean'), ascending = False)
highest_pm25_concentration = highest_pm25_concentration[['pm2.5_atm']]
highest_pm25_concentration.head(5)
```

```python
#isolating pm10.0 top 5 locations
highest_pm10_concentration = concentration_summary.sort_values(by = ('pm10.0_atm', 'mean'), ascending = False)
highest_pm10_concentration = highest_pm10_concentration[['pm10.0_atm']]
highest_pm10_concentration.head(5)
```

#Task 2

```python
stat_summary = air_data.groupby(['sensor.name','date']).agg({
    'pm2.5_atm':['max', 'min', 'mean', 'median'],
    'pm10.0_atm':['max', 'min', 'mean', 'median'],
    'voc':['max', 'min', 'mean', 'median']})
stat_summary = stat_summary.sort_values(by=('voc','max'),ascending = False)

#stat_summary.head(5)
pivot_stats2 = pd.pivot_table(
    stat_summary,
    index =["sensor.name",'date'],
    values = [("pm2.5_atm",'max'), ("pm10.0_atm",'max'),("voc",'max')])

pivot_stats2

```

```python
pivot_stats2 = pd.pivot_table(
    stat_summary,
    index =["sensor.name"],
    values = [("pm2.5_atm",'max'), ("pm10.0_atm",'max'),("voc",'max')])
    #values = [("pm2.5_atm",'max'), "pm10.0_atm","voc"])
pivot_stats2.head
```

#Task 3

```python


#step 1: define a function that assigns temperature levels 
def assign_temperature(temperature): 
    if temperature < 32: 
        return "Below Freezing" 
    elif 32<= temperature <= 50: 
            return "Cool" 
    elif 51 <= temperature <= 70: 
        return "Warm" 
    else: 
        return "Hot"
        
def assign_humidity(humidity): 
    if humidity < 50: 
        return "Low" 
    elif 51<= humidity <= 80: 
        return "High" 
    else: 
        return "Very High" 
        
air_data["temperature_level"] = air_data["temperature"].apply(assign_temperature) 
air_data["humidity_level"] = air_data["humidity"].apply(assign_humidity) 
air_data.head(5)

 
    
#stat_summary
 
#Pivot table
pivot_stats3 = pd.pivot_table(
    air_data,
    index = ['temperature_level','humidity_level'],
    values = ['pm2.5_atm','pm10.0_atm'],
    aggfunc = ['mean','median','max','min']
)
pivot_stats3
```

#Task 4

```python
#sorting by pm2.5 and pm10.0 limits
unhealthy_pm25 = air_data[air_data['pm2.5_atm']>=35.4]
unhealthy_events = unhealthy_pm25[unhealthy_pm25['pm10.0_atm']>=154]

#only keeping relevant data
unhealthy_events = unhealthy_events[
    [ 'sensor.name','date', 'pm2.5_atm','pm10.0_atm'] 
].sort_values(by = 'pm2.5_atm',ascending = False)
unhealthy_events
```
